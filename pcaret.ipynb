```python
import pandas as pd
import numpy as np
```

```python
# load processes features
feature_df = pd.read_csv('model/train/output/feature_set/feature_set.csv')
feature_df = feature_df.drop(columns=['Unnamed: 0'])
```

```python
from pycaret.classification import *
```

```python
s = setup(feature_df, target='label', train_size=.8, normalize=True, transformation=True, create_clusters=True, data_split_shuffle=True, data_split_stratify=True)
```

```python
best3 = compare_models(n_select=3, fold=20)
```

```python
blened_model = blend_models(best3, fold=20)
```

```python
stacked_model = stack_models(best3, fold=20)
```

```python
model = finalize_model(stacked_model)
```

```python
save_model(model, 'model')
```
